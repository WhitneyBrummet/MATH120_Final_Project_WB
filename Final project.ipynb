```python
class Reader:
    """Reader: This class has 2 attributes:
        -- file: A string that translates to a file name.
        -- List: A list initialized as empty but can be transformed to a list of words read from the file."""
    def __init__(self, file):
        """__init__: This function initializes a reader object with the file attribute assigned to the file name given and List
        attribute assigned as initially empty. If file is not a txt file, the function will ask again for the file name."""
        while file[-4:] != ".txt":
            print("File name not valid!")
            file = input("Please enter valid file name: ")
        self.file = file
        self.List = []
    def read(self):
        """read: This function takes the object's attribute 'file', opens the file with that name, reads the file, and splits
        the words of the file into a list and returns that list."""
        with open(self.file, "r") as fle:
            lines = fle.readlines()
            self.List = lines[0].replace('"', "")
            self.List = self.List.split(",")
        return self.List
```
